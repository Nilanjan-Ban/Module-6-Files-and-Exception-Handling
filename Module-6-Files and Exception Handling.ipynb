```python
'''15.How does Python's garbage collection system work?

Ans:

Python's garbage collection system is responsible for automatically managing memory by identifying and removing objects that are no longer in use, thereby preventing memory leaks and optimizing resource usage. Here’s an overview of how Python’s garbage collection works:

1. Reference Counting
Reference Counting is the primary mechanism used by Python to track objects in memory.
Each object in Python has an associated reference count, which tracks the number of references to that object.
 When we create a reference to an object, its reference count increases. When a reference is deleted or goes
 out of scope, the reference count decreases.
When an object’s reference count reaches zero (meaning no references to it remain), the memory occupied by
the object is automatically deallocated.


2. Cycle Detection
While reference counting works well for many cases, it fails when objects reference each other
(circular references). For example, if two objects refer to each other, their reference counts might never
reach zero, even though they are no longer in use.
Python handles this problem with cyclic garbage collection. It periodically checks for objects that
reference each other in cycles and can safely be deleted.

3. The Garbage Collector (GC)
Python uses a generational garbage collection system that works in cycles. The idea is to divide objects into generations based on their age:
Generation 0: New objects are initially placed in this generation.
Generation 1: Objects that survive one garbage collection cycle are promoted to this generation.
Generation 2: Objects that survive multiple garbage collection cycles are promoted to the oldest generation.
Young objects (generation 0) are collected more frequently because they are more likely to become garbage quickly. Older objects (generations 1 and 2) are collected less frequently to minimize overhead.
This generational approach reduces the performance cost of garbage collection, as it avoids scanning objects that have been in memory for a long time and are unlikely to be garbage.

4. Manual Control
Python allows you to interact with the garbage collector using the gc module:

Trigger manual garbage collection: You can force a garbage collection cycle with gc.collect().
Control collection behavior: We can configure the frequency and behavior of the collector using gc.set_threshold().


5. __del__ Method
If you define a __del__ method in a class, Python will attempt to call it when an object is about to be destroyed (e.g., when its reference count reaches zero).
However, relying solely on __del__ for resource management can be problematic, especially with circular references, since it may delay the cleanup process.

'''
```

```python
'''23.What are the challenges associated with memory management in Python?

Ans: Memory management in Python comes with several challenges that developers need to consider to ensure efficient use of system resources and optimal performance. Some of the key challenges include:

Garbage Collection (GC) Overhead: Python uses automatic memory management with a garbage collector to reclaim unused memory, particularly for cyclic references (objects referring to each other). While this simplifies development, the garbage collector can introduce overhead, especially in large programs, leading to occasional performance hiccups when it runs. Developers may not have direct control over when garbage collection happens, which can affect performance unpredictably.

Reference Counting and Circular References: Python uses reference counting as the primary mechanism to track object lifetimes. Each object has a reference count, which increases when a reference to the object is made and decreases when a reference is deleted. When the reference count drops to zero, the memory is freed. However, circular references (where objects reference each other in a cycle) can prevent the reference count from reaching zero, leading to memory leaks unless the garbage collector detects and breaks the cycle.

Memory Fragmentation: While Python handles memory allocation through its internal memory pool (using the pymalloc allocator), this can lead to fragmentation. If many small objects are created and deleted frequently, it might cause inefficient memory usage and performance slowdowns.

Object Overhead: Every object in Python carries overhead due to its internal data structures. For example, even a simple integer in Python is an object with a reference count and type information, which consumes more memory compared to lower-level languages like C. This can be a concern when dealing with large numbers of small objects.

Memory Leaks: Even with garbage collection, memory leaks can still occur in Python. This happens if objects are unintentionally held in memory due to lingering references (such as references stored in global variables, closures, or data structures). Identifying and eliminating such leaks can be difficult, especially in complex applications.

Large Objects and Memory Consumption: Some operations, such as dealing with large datasets (e.g., in scientific computing or machine learning), can consume a lot of memory. Python’s memory model can lead to higher memory consumption compared to other languages, especially when creating large lists, arrays, or data structures, as Python objects carry more overhead.

Interfacing with External Libraries: When interacting with C extensions, third-party libraries, or low-level APIs, Python’s memory management may conflict with the memory management models of these external systems. For instance, if an external library allocates memory but doesn’t properly release it, Python’s garbage collector may not be able to free it.

Memory Management for Multithreading: While Python's Global Interpreter Lock (GIL) makes it difficult to manage memory in multi-threaded applications efficiently, managing memory across threads can still be challenging, particularly when each thread maintains its own set of objects or when objects are shared between threads.
'''
```

```python
#9.How can you read a file line by line and store its content in a list in Python?

# Open the file in read mode
file_name = input("Enter the file name to read: ")
lines = []  # List to store file lines

try:
    with open(file_name, "r") as file:
        # Read the file line by line and store each line in the list
        for line in file:
            lines.append(line.strip())  # strip() removes any leading/trailing whitespace (including newline characters)
    print("File content stored in list:")
    print(lines)
except FileNotFoundError:
    print("Error: The file does not exist.")
except PermissionError:
    print("Error: You do not have permission to read the file.")
```

```python
#12.Write a program that demonstrates using multiple except blocks to handle different types of exceptions?

def handle_exceptions():
    try:
        # User input for a division operation
        num1 = int(input("Enter the numerator: "))
        num2 = int(input("Enter the denominator: "))
        result = num1 / num2
        print(f"Result of division: {result}")

        # Trying to open a file
        file_name = input("Enter the name of the file to open: ")
        with open(file_name, 'r') as file:
            content = file.read()
            print("File content:\n", content)

    except ZeroDivisionError as e:
        print(f"Error: Division by zero is not allowed. {e}")

    except ValueError as e:
        print(f"Error: Invalid input. Please enter valid numbers. {e}")

    except FileNotFoundError as e:
        print(f"Error: The file was not found. {e}")

    except Exception as e:
        # Generic exception handler for any other unexpected errors
        print(f"An unexpected error occurred: {e}")

# Run the function
handle_exceptions()
```

```python
#15.Write a Python program that prints the content of a file and handles the case when the file is empty

def print_file_content(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            if content:
                print(content)
            else:
                print("The file is empty.")
    except FileNotFoundError:
        print(f"The file at {file_path} does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file_path = "example1.txt"
print_file_content(file_path)
```

```python
#16.Demonstrate how to use memory profiling to check the memory usage of a small program

import logging

def setup_logger():
    logging.basicConfig(
        filename='error.log',
        level=logging.ERROR,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

def read_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except FileNotFoundError:
        logging.error("File not found: %s", filename)
        print("Error: The specified file was not found.")
    except Exception as e:
        logging.error("An unexpected error occurred: %s", str(e))
        print("An unexpected error occurred.")

# Setup logger
setup_logger()

# Example usage
filename = "example.txt"
content = read_file(filename)
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB

import logging
from logging.handlers import RotatingFileHandler

# Configure logging
log_file = "app.log"
log_formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")

handler = RotatingFileHandler(log_file, maxBytes=1_000_000, backupCount=5)  # 1MB per file, keep 5 backups
handler.setFormatter(log_formatter)

logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Example log messages
for i in range(10):
    logger.info(f"Log entry {i}")
```

```python
#19.Write a program that handles both IndexError and KeyError using a try-except block

def handle_errors():
    my_list = [1, 2, 3]
    my_dict = {'a': 10, 'b': 20}

    try:
        # Trying to access an out-of-range index
        print(my_list[5])

        # Trying to access a non-existent key in the dictionary
        print(my_dict['c'])
    except IndexError:
        print("IndexError: Tried to access an index that does not exist.")
    except KeyError:
        print("KeyError: Tried to access a key that does not exist in the dictionary.")

handle_errors()
```

```python
#21. Write a Python program that reads a file and prints the number of occurrences of a specific word
def count_word_occurrences(filename, word):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
            words = content.split()
            count = words.count(word)
            print(f"The word '{word}' occurs {count} times in the file.")
    except FileNotFoundError:
        print("Error: The specified file was not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
filename = "example.txt"
word_to_count = "appending"
count_word_occurrences(filename, word_to_count)
```
